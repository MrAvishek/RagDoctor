```python
from langchain_community.document_loaders import PyPDFLoader, DirectoryLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
#extract data from pdf
def load_pdf_data(path):
  loader = DirectoryLoader(path, glob="*.pdf", loader_cls=PyPDFLoader)
  documents = loader.load()
  return documents
```

```python
corpus_dir= os.path.join(curr_dir,'..','book')
```

```python
extracted_text = load_pdf_data(path=corpus_dir)
```

```python
print(extracted_text[4])
```

```python
#split text into chunks
def text_splitter(extracted_data):
  text_splitter = RecursiveCharacterTextSplitter(
      chunk_size = 500,
      chunk_overlap  = 50,
      length_function = len,
  )
  text_chunk= text_splitter.split_documents(extracted_data)
  return text_chunk
```

```python
text_chunk = text_splitter(extracted_text)
```

```python
# checking if the name is present or not
def checkVectorDB(name,PCapi):
    pc = Pinecone(api_key=PCapi)
    existing_indexes = pc.list_indexes()
    for i in range(len(existing_indexes)-1):
        if existing_indexes[0]['name'] == name:
            return True
    else:
        return False
    
```

```python
name = 'doctordb'
checkVectorDB(name)
```

```python
pc = Pinecone(api_key=PCapi)
index_name = "doctordb"
if not checkVectorDB(index_name,PCapi):
    pc.create_index(
        name=index_name,
        dimension=384, #  model dimensions
        metric="cosine", #model metric
        spec=ServerlessSpec(
            cloud="aws",
            region="us-east-1"
        )
    )
else:
    print("database created Previously!")
```

```python
from langchain.chains import RetrievalQA
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
# from google.colab import userdata
gemapi= os.getenv('GemAPI')
# GemAPI = userdata.get('GemAPI')
os.environ["GOOGLE_API_KEY"] = gemapi
```

```python
llm = ChatGoogleGenerativeAI(
    model="gemini-2.0-flash",  # or "gemini-1.5-flash"
    temperature=0.7,
    max_tokens=1024
)
```

```python
from langchain.prompts import PromptTemplate
custom_prompt = PromptTemplate(
    input_variables=["context", "question"],
    template="""
You are an expert in holistic and alternative medicine, with deep knowledge sourced from trusted resources like *The Gale Encyclopedia of Alternative Medicine*.

Using the following context from the encyclopedia, answer the question thoughtfully and factually. Focus on natural remedies, therapies, and traditional practices when relevant.

If the answer cannot be found in the context, respond with: "The provided encyclopedia content does not contain a definitive answer."
“Use a calm, educational tone.”

Context:
{context}

Question:
{question}

Answer:
"""
)
```

```python
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=retriver,
    return_source_documents=True,
    chain_type_kwargs={"prompt": custom_prompt}
)
```

```python
query = "what is Atherosclerosis why it happens"
result = qa_chain.invoke(query)

print("Answer:", result['result'])
# print("\nSource Documents:")
# for doc in result['source_documents']:
    # print(doc.metadata.get('source', 'Unknown Source'))
```
